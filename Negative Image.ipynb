```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
orignal=plt.imread("dog.png")
```

```python
orignal=orignal[:,:,0]
```

```python
negative=255-orignal
```

```python
fig,axes=plt.subplots(1,2,figsize=(8,9),dpi=200)
axes[0].set_title("Orignal IMAGE")
axes[0].imshow(orignal,cmap="gray")
axes[1].set_title("Negative IMAGE")
axes[1].imshow(negative,cmap="gray")
```

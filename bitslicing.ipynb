```python
def bitslice(x,l):
    p=['0','00','000','0000','00000','000000','000000','0000000']
    y=np.uint8(np.zeros(x.shape[0]))
    for i in range(x.shape[0]):
        y[i]=int(bin(x[i])[2:][:-l]+p[l-1],2)
    return y
```

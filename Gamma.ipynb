```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def gammaTransformation(image,constant,gamma):
    image=constant*np.power(image,gamma)
    return image
```

```python
image=plt.imread("dog.png")
image=image[:,:,0]
```

```python
Y=gammaTransformation(image,1,3)
```

```python
Y.shape
```

```python
fig,axes=plt.subplots(1,2,figsize=(7,7),dpi=200)
axes[0].imshow(image,cmap="gray")
axes[1].imshow(Y,cmap="gray")
```

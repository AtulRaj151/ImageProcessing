```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def PWLT(image,S,R):
    m=[]
    Y=np.zeros(image.shape)
    for i in range(len(S)-1):
        m=(S[i]-S[i+1])/(R[i]-R[i+1])
        Y[image<=R[i+1]]=(image[image<=R[i+1]])*m
    return Y
        
        
    
    
    
```

```python
S=[0,20,90,255]
R=[0,50,130,255]
```

```python
Y=np.uint8(PWLT(image,S,R))
Y.shape
```

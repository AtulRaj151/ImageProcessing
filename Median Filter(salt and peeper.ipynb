```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img=plt.imread("salt.png")
```

```python
def salt_peeper(image):
    m,n=image.shape
    f=3
    s=1
    Y=np.zeros(image.shape)
    for i in range(m):
        for j in range(n):
            ver_st=s*i
            ver_end=ver_st+f
            hor_st=j*s
            hor_end=hor_st+f
            img_slice=image[ver_st:ver_end,hor_st:hor_end]
            x=np.median(img_slice)
            img_slice[1:2,1:2]=x
            Y[ver_st:ver_end,hor_st:hor_end]=img_slice
    return Y
            
            
    
```
